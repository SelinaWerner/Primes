```python
import math
import matplotlib.pyplot as plt
import numpy as np
def isPrime(z):
    i = 2
    while i <= math.sqrt(z):
        if z % i == 0:
            return False
        i = i + 1
    return True

print(isPrime(2))
print(isPrime(4))
print(isPrime(5))
    
```

```python
def findDividors(z):
    list = []
    for i in range(2, z):
        if z%i == 0:
            list.append(i)
            list.append(int(z/i))
            break
    return list
```

```python
def Prim(z):
    list = [z]
    i = 0
    while True:
        if not isPrime(list[i]):
            list.extend(findDividors(list[i]))
            list.pop(i)
        else:
            i = i+1
            if i >= len(list):
                break
    return list

print(Prim(34252))
print(Prim(123423))
```

```python
 #Calculates how many primes there are that are smaller than n
def PrimCounter(n):
    pi_n = 0 
    for i in range(2, n+1):
        if isPrime(i): #isPrime from Implentation of Prime 
                        #factorisation (Bem 1.2)
            pi_n += 1 
    return pi_n
```

```python
#plot approximations and exact function
for n in range (2,4):    
    #arrays for exact step function
    a=[]
    b=[]
    for i in range (0,10**n):
        a.append(i)
        b.append(PrimCounter(i))

    # 100 linearly spaced numbers
    x = np.linspace(1.1,10**n,100)

    # setting the axes at the centre
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    #Gau√ü
    y=x/np.log(x)
    #Legendre
    z=x/(np.log(x)-1.08366)
    # plot the function
    plt.plot(x,z, 'g')
    plt.plot(x,y, 'r')
    plt.step(a, b, where='post')
    # show the plot
    plt.show()
```
